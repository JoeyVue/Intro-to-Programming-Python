```python
# Programming exercises
# Q1

def main():
    
    # Create list for days of the week.
    week_sales = []
    
    # list the the days of the week.
    for days in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']:
        
        # create float for user's input.
        sales = float(input(f"Sales for day {days}: $"))
        week_sales.append(sales)
        
    # Calculate the total of week sales.
    total = sum(week_sales)
    
    # Display the total amount of sales.
    print(f"Total amount of sales this week: ${total: .2f}")
    
main()
```

```python
# Q14
def main():
    
    # input the txt categories
    categories = {}
    expenses = []

    # import the matpltlib.pyplot
    import matplotlib.pyplot as plt
    
    # Read the expense data from txt
    with open('expenses.txt', 'r') as file:
        for line in file:
            category, expense = line.strip().split(',')
            categories[category] = float(expense)
            expenses.append(float(expense))
            
    # create pie chart
    plt.figure(figsize=(6,6))
    plt.pie(expenses, labels = categories.keys(), autopct ='%1.1f%%', startangle = 120)
    plt.title("Last month expenses")
    plt.axis('equal')
    plt.show
    
main()
```

```python
# Q2

def main():

    import random
    
    # create list for random numbers.
    lottery_numbers = []
    
    # create generate range for numbers.
    for _ in range(7):
        # create random number range
        random_number = random.randint (1,99)
        lottery_numbers.append (random_number)
        
    # display the random numbers.
    print("The lottery numbers are:")
    for number in lottery_numbers:
        print(number, end = ' ')
        
main()    
```
