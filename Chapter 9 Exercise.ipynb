#### Write a program that keeps names and email addresses in a dictionary as key-value pairs. The program should display a menu that lets the user look up a person’s email address, add a new name and email address, change an existing email address, and delete an existing name and email address. The program should pickle the dictionary and save it to a file when the user exits the program. Each time the program starts, it should retrieve the dictionary from the file and unpickle it.When testing this program, use fake names and email addresses.    

```python
# Q8


# Import the pickle module for data serialization.
import pickle
# import the os module for file operations.
import os

# load the dictionary from a file.
def load_data(file_name):
    # Check if the data file exists.
    if os.path.exists(file_name):
        # Open the file in binary read mode
        with open(file_name, 'rb') as file:
            # Deserialize data from the file
            data = pickle.load(file)
        return data
    else:
        # return an empty dictionry.
        return {}

#save the dictionary to a file.
def save_data(data, file_name):
    # open the file in binary write mode.
    with open(file_name, 'wb') as file:
        # Serialize and save data.
        pickle.dump(data, file)

def main():
    # Define the file name for storing data.
    data_file = 'email_data.pkl'
    #Load the email data from the file.
    email_dict = load_data(data_file)

    # Create prints for inputs.
    while True:
        print("\nEmail Address dictionary Menu:")
        print("1. Look up an email address")
        print("2. Add a new name and email address")
        print("3. Change an existing email address")
        print("4. Delete an existing name and email address")
        print("5. Exit")

        choice = input("Enter your choice (1/2/3/4/5): ")

        if choice == '1':
            # Search email's name.
            name = input("Enter the name to look up: ")
            if name in email_dict:
                print(f"Email Address for {name}: {email_dict[name]}")
            else:
                print(f"{name} not found.")

        elif choice == '2':
            # Create name for email
            name = input("Enter the new name: ")
            # Create email address.
            email = input("Enter the email address: ")
            email_dict[name] = email
            print(f"{name}'s email address added.")

        elif choice == '3':
            name = input("Enter the name to change the email address: ")
            # change email adress.
            if name in email_dict:
                email = input(f"Enter the new email address for {name}: ")
                email_dict[name] = email
                print(f"{name}'s email address has been updated.")
            else:
                print(f"{name} not found.")

        elif choice == '4':
            name = input("Enter the name to delete: ")
            # delete email data.
            if name in email_dict:
                del email_dict[name]
                print(f"{name}'s data has been deleted.")
            else:
                print(f"{name} not found.")

        elif choice == '5':
            # save email data.
            save_data(email_dict, data_file)
            print("Address saved.")
            break

        else:
            print("Invalid choice. Please choose number 1-5.")

# Run the main function            
if __name__ == "__main__":S
    main()
```

I started with "pickle" for serialzation and deserialization of data. Then I made the load and save data funtions to create and load existing files. I then created a main funtion to set the name and email files. Then I created a "while" loop for the user's interactions of 1-5 of the given options. The given options can apply, update, or delete entries. I finalized with "if __name__ == "__main__":" to execute the script.

#### Create A flowchart that connects to at least 3 people.

```python
# Import the NetworkX library for graph operations.
import networkx as nx
# Import Matplotlib for graph visualization.
import matplotlib.pyplot as plt
# Define a function to create a simple graph.
def create_simple_graph():
    # Create an undirected graph
    G = nx.Graph()

    # add people to graph.
    G.add_node("Person A")
    G.add_node("Person B")
    G.add_node("Person C")
     
    # Add edges to connect people.
    G.add_edge("Person A", "Person B")
    G.add_edge("Person B", "Person C")

    return G
# Define a function to draw the graph.
def draw_graph(graph):
    
    # Layout graph using spring layout algorithm.
    pos = nx.spring_layout(graph)
    
    #draw the graph with labels, node attributes, and styling.
    nx.draw(graph, pos, with_labels=True, node_size= 2500, node_color='pink' , font_size=11, font_color='black')
    # Show the graph using Matplotlib
    plt.show()

# Check if this script's main program.
if __name__ == "__main__":

    # Create a sumple graph.
    graph = create_simple_graph()
    
    # Draw graph.
    draw_graph(graph)
```

I decided to create my flowchart with simplegraph. So I started with importing the libraries for the graphs and visualization. I then created a funtion call "create_simple_graph" to create a simple undirectd graph. I then added people with "nodes" and then added "edges" to connect them. I defined the funtion with "draw_graph" so the graph will appear. nx.spring_layout was used to determine the layout of the graph. Then I used "nx.draw()" to customize the layout of the graph. The graph is then displayed with "plt.show()" and "graph = create_simple_graph()" creates the graph.
